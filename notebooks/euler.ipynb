```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the value of r
r = 1

# Generate a range of n values
n_values = np.arange(1, 1001)

# Calculate the sequence (1 + r/n)^n for each value of n
sequence_values = (1 + r/n_values)**n_values

# Calculate e^r
e_r = np.exp(r)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(n_values, sequence_values, label=r'$\left(1 + \frac{r}{n}\right)^n$', color='blue')
plt.axhline(y=e_r, color='red', linestyle='--', label=r'$e^r$')

# Annotate the plot
plt.title(r'Visualization of $\lim_{{n \to \infty}} \left(1 + \frac{r}{n}\right)^n = e^r$', fontsize=14)
plt.xlabel('n', fontsize=12)
plt.ylabel(r'$\left(1 + \frac{r}{n}\right)^n$', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```
