```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Base ice cream consumption pattern for one year (fluctuates around 0.4 gallons per month per capita)
base_consumption_gallons = np.array([0.2, 0.3, 0.35, 0.4, 0.5, 0.6, 0.8, 0.7, 0.55, 0.45, 0.35, 0.2])

# Generate 10 years of ice cream consumption with slight variability (in gallons per capita per month)
ice_cream_consumption_gallons = np.tile(base_consumption_gallons, 10) + np.random.normal(0, 0.02, 120)

# Simulate murders with a positive correlation to ice cream consumption, averaging around 60 per month
murders = 60 + (ice_cream_consumption_gallons - np.mean(ice_cream_consumption_gallons)) * 40
murders += np.sin(np.linspace(0, 12 * np.pi, 120)) * 8  # Add smoother seasonal variation
murders += np.random.normal(0, 5, 120)  # Minor random noise

# List of month names
month_labels = np.tile(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"], 10)

# Create a time axis for the 10 years
months = np.arange(1, 121)

# ---- Figure 1: Scatter Plot with Only Dots ----
plt.figure(figsize=(10, 6))
plt.scatter(ice_cream_consumption_gallons, murders, color='blue', s=100, alpha=0.7)
plt.title('Murders vs. Ice Cream Consumption (Only Dots)', fontsize=16)
plt.xlabel('Ice Cream Consumption (Gallons per Capita)', fontsize=12)
plt.ylabel('Number of Murders', fontsize=12)
plt.grid(True)
plt.show()

# ---- Figure 2: Time Series Plot with Two Y-axes ----
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot ice cream consumption on the left y-axis
ax1.set_xlabel('Months (1 to 120)')
ax1.set_ylabel('Ice Cream Consumption (Gallons per Capita)', color='blue')
ax1.plot(months, ice_cream_consumption_gallons, color='blue', label="Ice Cream Consumption")
ax1.tick_params(axis='y', labelcolor='blue')

# Create the second y-axis for murders, sharing the same x-axis
ax2 = ax1.twinx()
ax2.set_ylabel('Number of Murders', color='red')
ax2.plot(months, murders, color='red', label="Murders")
ax2.tick_params(axis='y', labelcolor='red')

# Add a title and grid
plt.title('Murders and Ice Cream Consumption Over 10 Years (Time Series)', fontsize=16)
plt.grid(True)

# Show the plot with two y-axes
plt.show()

# ---- Figure 3: Scatter Plot with Month Labels ----
plt.figure(figsize=(10, 6))
plt.scatter(ice_cream_consumption_gallons, murders, color='blue', s=100, alpha=0.7)

# Add month labels to the points
for i in range(len(month_labels)):
    plt.text(ice_cream_consumption_gallons[i], murders[i], month_labels[i], fontsize=9, ha='right')

# Set title and labels
plt.title('Murders vs. Ice Cream Consumption with Month Labels', fontsize=16)
plt.xlabel('Ice Cream Consumption (Gallons per Capita)', fontsize=12)
plt.ylabel('Number of Murders', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```
