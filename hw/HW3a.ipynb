```python
import numpy as np
import matplotlib.pyplot as plt

print("#1")
# 1)
treespermark2 = 20
ans1 = treespermark2*4.3/(180)**2
print("{:.2e}".format(ans1))
print("\n")

print("#2")
# 2a)
MWmassSolar = 2.06e11
planetPct = .125
ans2a = MWmassSolar/planetPct
print("{:.2e}".format(ans2a))

# 2b)
Sunmasskg = 1.988416e30
ans2b = MWmassSolar*Sunmasskg
print("{:.2e}".format(ans2b))
print("\n")

print("#3")
# 3)
from math import log10, floor
def round_sigfig(x, numsigfig=2):
    return round(x, numsigfig-int(floor(log10(abs(x))))-1)
# a)
arith1 = round_sigfig(9.18e5+.00821,3)
print("{:.2e}".format(arith1))
# b)
arith2 = round_sigfig(78.934*23.98,4)
print("{:.3e}".format(arith2))
# c)
arith3 = round_sigfig(8.4562e-5 + 1.34e3,3)
print("{:.2e}".format(arith3))
print("\n")

print("#4")
# 4)
# a)
m_ant = 1e-3  # Google search
m_human = 7e4  # 70kg closest order of magnitude for human
print("{:.2e}".format(m_human/m_ant))
print("~8 orders of magnitude.")

# b)
m_atom = 1.7e-27  # Google search
m_soccer = 4.5e2  # Google search
print("{:.2e}".format(m_soccer/m_atom))
print("~29 orders of magnitude.")

# c)
m_car = 1500e3  # Google search
m_MtE = 175e15  # https://mounteverest.info/mount-everest-weigh-mass-volume/
print("{:.2e}".format(m_MtE/m_car))
print("~11 orders of magnitude.")
print("\n")

# 5) 
# Intensive Properties
# Temperature does not depend on the amount of air or the size of the balloon. Whether the balloon is large or small, the temperature can be the same.
# Pressure, similar to temperature>
# Extensive Properties
# Volume depends on the size of the balloon. A larger balloon will have a greater volume of air compared to a smaller one.
# Mass varies with the amount of air present. A larger balloon will contain more air, thus having a greater mass.

print("#6")
# 6)
"""
This program asks the user to input a car's speed in km/hr; it converts the speed
to miles/sec and prints the converted speed to the screen.
Input and output are specified to three significant figures.

__author__      = "D.M. Ruth"
__date__        = "9/20/2024"
"""
# Ask the user for the speed in km/hr
kmph = float(input("Enter the car's speed in km/hr (to 3 significant figures): "))

# Convert the speed to miles/sec
# 1 hr = 3600 sec; 1000m = 1 km; 3 ft = 1 yd = 0.9144 m, 1 mi = 5280 ft
conversion_factor = (1/3600)*1000*(3/0.9144)*(1/5280)
mps = kmph * conversion_factor

# Display the result
print(f"The speed in miles/sec is: {mps:.2e}")
```
